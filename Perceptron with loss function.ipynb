```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,n_informative=1,n_classes=2,n_features=2,n_clusters_per_class=1,n_redundant=0,
                          hypercube=False,class_sep=15,random_state=41)
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,s=80)
plt.show()
```

```python
def perceptron(X,y):
    weights = np.ones(3)
    learning_rate = 0.1
    for j in range(1000):
        for i in range(X.shape[0]):
            fx = weights[0]*X[i][0] + weights[1]*X[i][1] + weights[2]
            if y[i]*fx < 0:
                weights[0] = weights[0] + learning_rate*y[i]*X[i][0]
                weights[1] = weights[1] + learning_rate*y[i]*X[i][1]
                weights[2] = weights[2] + learning_rate*y[i]
    return [weights[0],weights[1]],weights[2]
```

```python
coef,intercept = perceptron(X,y)
```

```python
coef
```

```python
intercept
```

```python
m = -(coef[0]/coef[1])
b = -(intercept/coef[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,s=80)
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.ylim(-2.5,2.5)
plt.show()
```
