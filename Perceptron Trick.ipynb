```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100,n_informative=1,n_classes=2,n_features=2,n_clusters_per_class=1,n_redundant=0,
                          hypercube=False,class_sep=2.3,random_state=13)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,s=80)
plt.show()
```

```python
def step(n):
    return 1 if n > 0 else 0
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    learning_rate = 0.1
    epochs = 1000
    for i in range(epochs):
        j = np.random.randint(0,100)
        y_pred = step(np.dot(X[j],weights))
        weights = weights + learning_rate*(y[j]-y_pred)*X[j]
    return weights[1:],weights[0]
```

```python
coef,intercept=perceptron(X,y)
```

```python
coef
```

```python
intercept
```

```python
m = -(coef[0]/coef[1])
b = -(intercept/coef[1])
```

```python
x_input = np.linspace(-1.5,2,100)
y_input = m*x_input + b
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y,s=80)
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.ylim(-3,2)
plt.show()
```
